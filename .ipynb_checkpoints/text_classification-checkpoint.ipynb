# Basic preprocessing


### Stopwords removal

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words
```

```python
df['text'] = df['News'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
print(len(df['text'][0]))
```

```python
print(df['text'])
```

## Porter Stemming


```python
ps = PorterStemmer()
```

```python
df['text'] = df['text'].apply(lambda x: ' '.join([ps.stem(word) for word in x.split()]))
```

```python
print(df['text'])
```

### Converting to lowercase


```python
df['text'] = df['text'].apply(lambda x: ' '.join(x.lower() for x in x.split()))
```

```python
df['text'][0]
```

### Removing Punctuation


```python
df['cleaned'] = df['text'].str.replace('[^\w\s]','')
```

## Model Selection

### Splitting dataset

```python
X = df['cleaned']
Y = df['category']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=101)
```

### Feature Generation

```python
tfidf_vectorizer = TfidfVectorizer(min_df= 3, stop_words="english", norm='l2')
```

### Applying Algorithms

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC,LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics import classification_report, confusion_matrix

from sklearn.model_selection import cross_val_score

LR = LogisticRegression()
SVC = SVC()
NB = MultinomialNB()
```

```python
# Pipeline for logistic regression
LR_pipeline = Pipeline([('vect', tfidf_vectorizer),
                     ('chi',  SelectKBest(chi2, k=6000)),
                     ('clf', LogisticRegression(random_state=0))])
```

```python
# Pipeline for Support vector classifier
SVCpipeline = Pipeline([
    ('bow', tfidf_vectorizer),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', LinearSVC(C=6000)),  # train on TF-IDF vectors w/ SVM
])
```

```python
# Pipeline for Naive Bays Classifier
NB_pipeline = Pipeline([('tfidf', tfidf_vectorizer),
               ('clf', MultinomialNB()),
              ])
```

```python
###Training Models
NB_model = NB_pipeline.fit(X_train, y_train)
LR_model =  LR_pipeline.fit(X_train, y_train)
SVC_model =  SVCpipeline.fit(X_train, y_train)
```

### Model Selection

```python
from sklearn.metrics import accuracy_score

#Accuracy score of Naive Bays Classifier
NB_predict = NB_model.predict(X_test)
print("NB Test accuracy = ",accuracy_score(y_test,NB_predict))
print('\n')

#Accuracy score of Logistic Regression Classifier
LG_predict = LR_model.predict(X_test)
print("LG Test accuracy = ",accuracy_score(y_test,LG_predict))
print('\n')

#Accuracy score of Support Vector Classifier
SVC_predict = SVC_model.predict(X_test)
print("SVC Test accuracy = ",accuracy_score(y_test,SVC_predict))
print('\n')
```

```python
from sklearn import metrics
from matplotlib.pyplot import xticks
from matplotlib.pyplot import yticks
ytest = np.array(y_test)

# Confussion Matrix and confussion report of Naive Bays Classifier
print(classification_report(ytest, NB_model.predict(X_test)))
confusion_matrix = metrics.confusion_matrix(ytest, NB_model.predict(X_test))
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["International","blog","entertainment","political","social","sports","view"])


cm_display.plot()
xticks(rotation = 90)
yticks(rotation = 0)
plt.figure().set_figwidth(100)

plt.show()
```

```python
# Confussion Matrix and confussion report of Logistic Regression Classifier

ytest = np.array(y_test)
print(classification_report(ytest, LR_model.predict(X_test)))
confusion_matrix = metrics.confusion_matrix(ytest, LR_model.predict(X_test))
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["International","blog","entertainment","political","social","sports","view"])


cm_display.plot()
xticks(rotation = 90)
yticks(rotation = 0)
plt.figure().set_figwidth(100)

plt.show()
```

```python
# Confussion Matrix and confussion report of Support Vector Classifier

ytest = np.array(y_test)
print(classification_report(ytest, SVC_model.predict(X_test)))
confusion_matrix = metrics.confusion_matrix(ytest, SVC_model.predict(X_test))
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = ["International","blog","entertainment","political","social","sports","view"])


cm_display.plot()
xticks(rotation = 90)
yticks(rotation = 0)
plt.figure().set_figwidth(100)

plt.show()
```
